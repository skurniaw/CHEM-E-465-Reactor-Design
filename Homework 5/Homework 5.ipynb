```
# CHEM E 465 HW5 Sebastian K
# Question 3
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats
```

```
# Importing in our experimental data
CA = np.array([1.00, 1.29, 1.67, 2.15, 2.78, 3.59, 4.64, 5.99, 7.74, 10.00]);
ra = np.array([4.27, 5.36, 6.92, 8.06, 9.15, 11.4, 13.8, 15.0, 19.5, 19.3]) * 10**-3; # negative r_a
ra_i = np.array([2.13, 2.66, 3.22, 4.02, 4.71, 5.52, 6.51, 7.16, 8.62, 9.00]) * 10**-3; # negative r_a, inhibited
```

#**3a**

```
# determining MM parameters for non-inhibited case

slope, intercept, rval, pval, std_err = stats.linregress(CA**-1, ra**-1);

Vmax = intercept**-1;
KM = slope * Vmax;

print("For non-inhibited case, Vmax is: ", Vmax,". Michaelis-Menten constant is: ", KM)
```

#**3b**

```
plt.plot(CA**-1, ra**-1, label="Non-inhibited");
plt.plot(CA**-1, ra_i**-1, label="Inhibited");
plt.xlabel('1/C_A [L/mmol]')
plt.ylabel('-r_a [L min / (mmol)]')
plt.title("Lineweaver-Burk Plot for both inhibited and non-inhibited cases")
plt.legend()
```

Since both slope and intercept change, it is non-competitive inhibition.

#**3c**

```
slope1, intercept1, rval1, pval1, std_err1 = stats.linregress(CA**-1, ra_i**-1);

# rearrange the slope in terms of known variables to calculate Ki

CI = 0.1;
KI = (slope1 * (Vmax / KM) - 1)**-1 * CI;

print("KI: ", KI, "mol/L")
```
